```python
# Import all the god libs 
import pandas as pd
import numpy as np 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the csv file. You guys have to make sure that the csv file is in the same folder as the code.
DataFrame = pd.read_csv("whl_2025.csv") 
#Shows
pd.set_option('display.max_columns', None)
```

```python
DataFrame
```

```python
print("Before cleaning")
print(len(DataFrame))
```

```python
# Takes away any stupid duplicates. 
DataFrame = DataFrame.drop_duplicates() 
DataFrame
```

```python
#This adds a winner column.
DataFrame.loc[DataFrame['home_goals'] > DataFrame['away_goals'] , 'Winner'] = DataFrame['home_team'] 
DataFrame.loc[DataFrame['home_goals'] < DataFrame['away_goals'] , 'Winner'] = DataFrame['away_team'] 
```

```python
 DataFrame
```

```python
#Cleans the dataset, though in this case the data is very clean....
NoNa = DataFrame.dropna() 
NoNa
```

```python
#This seperates the winner countries from the losers. 

HomeTeam = NoNa[["record_id","Winner","toi","home_team","home_off_line","home_def_pairing","home_assists","home_shots","home_xg","home_max_xg","home_penalties_committed","home_penalty_minutes"]].copy()
HomeTeam.columns = ["id","Winner","toi","team","off_line","def_pairing","assists","shots","xg","max_xg","penalties_committed","penalty_minutes"]
HomeTeam["Home"] = 1 
HomeTeam["win"] = np.where(HomeTeam["Winner"] == HomeTeam["team"],1,0)
 
AwayTeam = NoNa[["record_id","Winner","toi","away_team","away_off_line","away_def_pairing","away_assists","away_shots","away_xg","away_max_xg","away_penalties_committed","away_penalty_minutes"]].copy()
AwayTeam.columns = ["id","Winner","toi","team","off_line","def_pairing","assists","shots","xg","max_xg","penalties_committed","penalty_minutes"]
AwayTeam["Home"] = 0 
AwayTeam["win"] = np.where(AwayTeam["Winner"] == AwayTeam["team"],1,0)

TeamData = pd.concat([HomeTeam,AwayTeam])
```

```python
#This is the one we are going to work with. 
TeamData
```

```python
sns.scatterplot(x = "toi",y = "penalties_committed", data = TeamData)
```

```python
sns.scatterplot(x = "toi",y = "xg", data = TeamData)
```

```python
sns.scatterplot(x = "penalties_committed",y = "win", data = TeamData)
```

```python
sns.barplot(x="penalties_committed", y="win", data=TeamData)
```
